```python
# make a plot wothout numpy in bar plot
nut_price = {"Almond butter":10,
             "Peanut butter":20,
             "Cashew butter":15}
fig,ax = plt.subplots()
ax.bar(nut_price.values(),nut_price.keys());
ax.set(title="Bar plot",
       xlabel = "price",
      ylabel="butter")
plt.show();
```

```python

heart = pd.read_csv("heart-disease.csv")
heart.head()
```

```python
# creating a histogram for age
heart["age"].plot.hist(bins=60);
```

```python
heart.plot.hist(figsize = (10,30), subplots='True');
```

```python
heart.head()
```

```python
over_50 = heart[heart["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',x='chol',y='age',c='target');
```

```python
#OO method with pyplot method
fix,ax = plt.subplots(figsize = (10,7))
over_50.plot(kind='scatter',x='chol',y='age',c='target',ax=ax);
#creating more space for the axis
#ax.set_xlim([45,100]);
```

## OO method from scratch

```python
fig,ax = plt.subplots(figsize=(10,6))
#plot it
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
#customize
ax.set(title="Heart disease and cholesterol levels",xlabel="age",ylabel="cholesterol");
#add legends
ax.legend(*scatter.legend_elements(),title="target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
#subplots for chol,thalach,age
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10)) 

#add data to ax0
scatter =ax0.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["chol"])

#customize
ax0.set(title="heart disease & cholesterol levels",xlabel="age",ylabel="thalach");

#adda a legend
ax0.legend(*scatter.legend_elements(),title="cholesterol");

#add a horizontal line
ax0.axhline(over_50["thalach"].mean(),linestyle="--")

#add data to ax1
scatter =ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])

#customize
ax1.set(title="heart disease & cholesterol levels",xlabel="age",ylabel="thalach");

#legend
ax1.legend(*scatter.legend_elements(),title="target");

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),linestyle="--")

#add title to the fig
fig.suptitle("Heart Disease Analysis",fontweight="bold",fontsize=16);
```
